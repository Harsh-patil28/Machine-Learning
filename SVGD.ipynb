# Single variable Linear Regression with Gradient Descent

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary.csv")
df.head()
```

```python
X = df["YearsExperience"].values
y = df["Salary"].values
```

```python
# Define the learning rate and the number of iterations
learning_rate = 0.001
num_iterations = 1000
```

```python
# Initialize the parameters (slope(1) and intercept(0))
theta0 = np.random.randn()
theta1 = np.random.randn()
```

```python
# Perform gradient descent
for i in range(num_iterations):
    # Calculate the predicted values
    y_pred = X*(theta1) + theta0
    
    # Calculate the error
    error = y_pred - y
    
     # Calculate the gradient of the cost function with respect to theta
    gradient_theta0= np.array(1 / len(X) * np.sum(error))
    gradient_theta1= np.array(1/ len(X) * np.sum(error * X))
    
       # Update the parameters using the gradient and learning rate
    theta0 = theta0 - learning_rate * gradient_theta0
    theta1 = theta1 - learning_rate * gradient_theta1
```

```python

# The final values of theta are the model's parameters
intercept, slope = theta0, theta1
print("Final intercept (theta0):", intercept)
print("Final slope (theta1):", slope)

# Plot the data and the regression line
plt.scatter(X, y, label="Data",color="green")
plt.plot(X, X.dot(slope) + intercept, color='red', linewidth=3, label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
