# Polynomial Regression

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)                                        # 0 means pseudo random numbers which repeats same 
X = np.sort(5 * np.random.rand(80,1),axis=0)
Y = np.sin(X).ravel()+np.random.rand(80)
```

```python
linear_reg = LinearRegression()
linear_reg.fit(X,Y)
```

```python
X_range = np.arange(0,5,0.1)[:,np.newaxis]
```

```python
degree = 3
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly,Y)
```

```python
X_range_poly = poly_features.transform(X_range)
Y_poly_pred = poly_reg.predict(X_range_poly)
```

```python
plt.scatter(X,Y,color = 'blue',label='Data')
plt.plot(X_range,linear_reg.predict(X_range),color='green',label="Linear Regression")
plt.plot(X_range,Y_poly_pred,color='red',label='Polynomial Regression(degree({}))'.format(degree))
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
